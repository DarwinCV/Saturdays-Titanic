## SibSp y Parch

```python
print(df.SibSp.unique())
print(df.Parch.unique())

# se suma 1 para corregir familias con 0 personas
df['Family'] = df['SibSp'] + df['Parch'] + 1

# inspect the correlation between Family and Survived
df[['Family', 'Survived']].groupby(['Family'], as_index=False).mean()
```

Se observa que para familias de 1 a 4 personas el porcentaje de sobrevivientes aumenta de manera proporcional. Al parecer para familias muy numerosas se convierte en un factor que disminuye su probabilidad de sobrevivencia.

```python
tb1 = df['Family'].value_counts().reset_index()
tb1.columns = ['Numero de personas', 'Count']
tb1['Porcentaje'] = 100*(tb1['Count']/ len(df))
tb1.sort_values(by=['Numero de personas'])
display(tb1)
```

El numero de familias con mas de 4 personas en aproximadamente el 5% y su porcentaje de sobrevivientes es menor al de familias de 1 persona por lo cual podemos fusionar las familias con mas de 4 personas en una categoria de '0' personas.

```python
df.Family = df.Family.map(lambda x: 0 if x > 4 else x)
df[['Family', 'Survived']].groupby(['Family'], as_index=False).mean()
```

## Modelamiento 

```python
df2 = df.drop(labels=['SibSp','Parch','Age','Fare','Cabin','Fare-bin2','Embarked','Title'], axis=1)
df2.head()
```

```python
y_train = df2[0:891]['Survived'].values
X_train = df2[0:891].drop(['Survived','PassengerId'], axis=1).values
X_test  = df2[891:].drop(['Survived','PassengerId'], axis=1).values
```

```python
# !pip install tensorflow

import keras 
from keras.models import Sequential 
from keras.layers import Dense
```

```python
model = Sequential()

# layers
model.add(Dense(9, kernel_initializer = 'uniform', activation = 'relu', input_dim = 17))
model.add(Dense(9, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(5, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# summary
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Train the NN
model.fit(X_train, y_train, batch_size = 32, epochs = 200)
# import sys
# import numpy
# numpy.set_printoptions(threshold=sys.maxsize)
# df
# print(X_train)
```
